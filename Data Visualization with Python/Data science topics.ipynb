```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
datascience_survey_data = pd.read_csv('https://cocl.us/datascience_survey_data',index_col=0)
datascience_survey_data
```

```python
# for sake of consistency, let's also make all column labels of type string
datascience_survey_data.columns = list(map(str, datascience_survey_data.columns))
```

```python
datascience_survey_data.sort_values(['Very interested'], ascending=False, axis=0, inplace=True)
datascience_survey_data
```

```python
percentage = (datascience_survey_data / 2233 *100).round(2)
percentage
```

```python
percentage.plot(kind='bar', figsize=(20, 8), width = 0.8)
```

```python
bars = percentage.plot(kind='bar', figsize=(20, 8), width = 0.8, color=('#5cb85c', '#5bc0de', '#d9534f'))


plt.legend(labels=percentage.columns,fontsize= 14)
plt.title("Percentage of Respondents' Interest in Data Science Areas",fontsize= 16)

plt.xticks(fontsize=14)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.yticks([])
```

```python
bars = percentage.plot(kind='bar', figsize=(20, 8), width = 0.8, color=('#5cb85c', '#5bc0de', '#d9534f')\
                ,y=['Very interested','Somewhat interested','Not interested'],fontsize=14)

plt.legend(labels=percentage.columns,fontsize= 14)
plt.title("Percentage of Respondents' Interest in Data Science Areas",fontsize= 16)

plt.xticks(fontsize=14)
plt.yticks([])

x=np.arange(len(percentage.index))
yv=np.array(list(percentage['Very interested']))
ys=np.array(list(percentage['Somewhat interested']))
yn=np.array(list(percentage['Not interested']))
for a,b in zip(x,yv): 
    plt.text(a-0.27,b+0.1,'%.2f'%b,ha = 'center',va = 'bottom',fontsize=14)
for a,b in zip(x,ys):
    plt.text(a,b+0.1,'%.2f'%b,ha = 'center',va = 'bottom',fontsize=14)
for a,b in zip(x,yn):
    plt.text(a+0.27,b+0.1,'%.2f'%b,ha = 'center',va = 'bottom',fontsize=14)
bars.spines['top'].set_visible(False)
bars.spines['right'].set_visible(False)
bars.spines['left'].set_visible(False)
plt.show
```
